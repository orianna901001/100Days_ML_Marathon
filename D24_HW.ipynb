下圖為一個機器學習分類器模型訓練經過50個 epoch 訓練後的結果。

在每一代的訓練中分別紀錄訓練集(Training)與驗證集(Validation)的損失曲線(cross entropy)和準確度(accuracy)。

試著說明看看這樣的訓練結果好不好？若結果不好可以怎樣改善提升模型的準確度？

訓練損失（Training Loss）不斷下降，訓練準確度（Training Accuracy）一路攀升，最終可達 0.9 以上，代表模型在訓練集上表現很好。

模型在驗證集上的表現遠不如訓練集，說明學到的特徵太過「記憶」訓練資料，無法泛化到新樣本。

Early Stopping：監控驗證損失，當連續 N 個 epoch（例如 5～10）未見改善，就自動停止訓練，並回到最佳權重。

Dropout：在全連接層或捲積層插入 dropout（0.3–0.5），隨機「關閉」部分神經元，迫使模型不過度依賴某些特徵。

擴充樣本多樣性，幫助模型學習更多樣性的特徵。
