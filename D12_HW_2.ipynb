```python
# 為學習方便，在此先移除遺失值
raw_data = raw_data.dropna(axis=1)
print(raw_data.shape)
```

```python
from sklearn.model_selection import train_test_split

trainData, testData = train_test_split(raw_data, test_size = 0.25, random_state = 214)
```

## Frequency Encoding

本次主要使用 Dictionary 資料型態的功能實現 Frequency Encoding，步驟如下：
1. 先計算各類別的數量
2. 將計算結果轉換成 Dictionary 資料型態
3. 進行類別轉換

```python
# Step1.、Step2. 先計算各類別的數量，且以 Dictionary 資料型態呈現
frequency=trainData['LandSlope'].value_counts().to_dict()

train_frequency=trainData['LandSlope'].apply(lambda x: frequency[x] if x in list(frequency.keys()) else 0)
test_frequency=testData['LandSlope'].apply(lambda x: frequency[x] if x in list(frequency.keys()) else 0)

# Step3. 進行類別轉換（特別注意沒有的類別要標記為 0）
```

```python
for before_frequency_encoding, after_frequency_encoding in zip(trainData["LandSlope"], train_frequency):
    print("Before Frequency Encoding: {}, After Frequency Encoding: {}".format(before_frequency_encoding, after_frequency_encoding))
```

## Feature Combination

本次將介紹使用 Dictionary 資料型態的功能進行類別特徵的特徵合併，其步驟如下：
1. 建構合併規則，並以 Dictionary 資料型態呈現
2. 將類別變數作轉換

舉例：Destination 與 VIP 兩個類別變數進行混合

```python
# Step1. 把 類別變數一 與 類別變數二 的類別取出來
one_category=trainData['LandSlope'].unique()
two_category=trainData['LotConfig'].unique()
# Step2. 建立一個二類別變數 Dict

two_dict={
    one_class:{
        two_class:"{}_{}".format(one_class,two_class) for two_class in two_category
    } for one_class in  one_category
}
# Step3. 在資料表中產生新特徵
trainData['LandSlope_LotConfig']=trainData.apply(lambda x:two_dict[x['LandSlope']][x['LotConfig']],axis=1)
print(trainData['LandSlope_LotConfig'])

```
