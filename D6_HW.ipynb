### 匯入所需套件

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 輸入資料

```python
import os
folder=r'C:\Users\Orianna\Desktop\data'
path=os.path.join(folder,'train.csv')
raw=pd.read_csv(path)
data=raw.copy()
data.head()
data.info()
```

### 繪製長條圖

```python
plt.rcParams['font.sans-serif']=['Microsoft JhengHei']
plt.bar(x=data['MSZoning'].value_counts().index,height=data['MSZoning'].value_counts())
plt.title('classification of the sale.')
plt.xlabel('MSZoning')
plt.ylabel('Frequency')
plt.show()
```

### 繪製盒鬚圖

```python
Built=data.loc[data["YearBuilt"].notna(),"SalePrice"]
Sold=data.loc[data["YrSold"].notna(),"SalePrice"]

plt.boxplot(x=[Built,Sold],labels=["Built","Sold"])
plt.show()
```

### 繪製散佈圖

```python
plt.scatter(x=data["OverallQual"],y=data["SalePrice"])

plt.show()
```

### 繪製折線圖

```python
data["OverallQual"]=pd.cut(data["OverallQual"],8,labels=["A","C","FV","I","RH","RL","RP","RM"])
Location_price=data.groupby(by="OverallQual").mean()["SalePrice"].to_frame().reset_index()
plt.plot(Location_price["OverallQual"],Location_price["SalePrice"])
plt.show()
```
